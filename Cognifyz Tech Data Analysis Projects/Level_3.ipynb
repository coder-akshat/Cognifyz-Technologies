```python
#Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import re
```

```python
#Reading the dataset
df = pd.read_csv("Dataset .csv")
df.head()
```

```python
#Getting more information regarding the dataset
df.describe()
```

```python
#Checking for null values
df.isna().sum()
```

```python
#Task 1:- Restaurant Reviews

#Analyze the text reviews to identify the most common positive and negative keywords.

#Calculate the average length of reviews and explore if there is a relationship between review length and rating.
```

```python
#Analyze the text reviews to identify the most common positive and negative keywords.

# Function to tokenize text into words
def tokenize_text(text):
    # Split text into words using regex 
    return re.findall(r'\w+', text.lower())

# Defining positive and negative keywords
positive_keywords = ['good', 'very good', 'great', 'excellent', 'awesome', 'delicious', 'tasty', 'fantastic']
negative_keywords = ['bad', 'poor', 'terrible', 'awful', 'disappointing', 'horrible', 'disgusting']

# Counter objects to store counts of positive and negative keywords
positive_counts = Counter()
negative_counts = Counter()

# Iterating over text reviews and count occurrences of positive and negative keywords
for review in df['Rating text']:
    words = tokenize_text(review)
    for word in words:
        if word in positive_keywords:
            positive_counts[word] += 1
        elif word in negative_keywords:
            negative_counts[word] += 1

# Getting the most common positive and negative keywords
most_common_positive = positive_counts.most_common()
most_common_negative = negative_counts.most_common()

print("Most common positive keywords:")
print(most_common_positive)

print("\nMost common negative keywords:")
print(most_common_negative)
```

```python
#Calculate the average length of reviews and explore if there is a relationship between review length and rating.
# Calculating the length of each review
df['Review Length'] = df['Rating text'].str.len()

# Calculating the average length of reviews
average_review_length = df['Review Length'].mean()
print("Average length of reviews:", average_review_length)

# Grouping the data by rating and calculate the average review length for each rating category
average_length_by_rating = df.groupby('Aggregate rating')['Review Length'].mean()
print("\nAverage review length by rating:")
print(average_length_by_rating)

# Analyze if there's a relationship between review length and rating
correlation = df['Aggregate rating'].corr(df['Review Length'])
print("\nCorrelation between review length and rating:", correlation)

# Checking the correalation between review length and rating
if correlation >= 0.5 and correlation <= 1:
    print("There is a strong positive coorelation between review length and rating")
elif correlation <= -0.5 and correlation >= -1:
    print("There is a strong negative coorelation between review length and rating")
else:
    print("There is no linear coorelation between review length and rating")
```

```python
#Task 2:- Votes Analysis

#Identify the restaurants with the highest and lowest number of votes.

#Analyze if there is a correlation between the number of votes and the rating of a restaurant.
```

```python
#Identify the restaurants with the highest and lowest number of votes.

# Sorting the restaurants by the number of votes in descending order
df_sorted_by_votes = df.sort_values(by='Votes', ascending=False)

# Restaurant with the highest number of votes
restaurant_highest_votes = df_sorted_by_votes.iloc[0]

# Restaurant with the lowest number of votes
restaurant_lowest_votes = df_sorted_by_votes.iloc[-1]

print("Restaurant with the highest number of votes:")
print(restaurant_highest_votes[['Restaurant Name', 'Votes']])

print("\nRestaurant with the lowest number of votes:")
print(restaurant_lowest_votes[['Restaurant Name', 'Votes']])
```

```python
#Analyze if there is a correlation between the number of votes and the rating of a restaurant.

# Calculating the Pearson correlation coefficient between rating and number of votes
correlation = df['Aggregate rating'].corr(df['Votes'])

print("Correlation between rating and number of votes:", correlation)

# Checking the correalation between rating and number of votes
if correlation >= 0.5 and correlation <= 1:
    print("There is a strong positive coorelation between the number of votes and the rating of a restaurant")
elif correlation <= -0.5 and correlation >= -1:
    print("There is a strong negative coorelation between the number of votes and the rating of a restaurant")
else:
    print("There is no linear coorelation between the number of votes and the rating of a restaurant")
```

```python
#Task 3:- Price Range vs. Online Delivery and Table Booking

#Analyze if there is a relationship between the price range and the availability of online delivery and table booking.

#Determine if higher-priced restaurants are more likely to offer these services.
```

```python
#Analyze if there is a relationship between the price range and the availability of online delivery and table booking.

# Creating a cross-tabulation of price range vs. online delivery
cross_tab_online_delivery = pd.crosstab(df['Price range'], df['Has Online delivery'], margins=True)

# Creating a cross-tabulation of price range vs. table booking
cross_tab_table_booking = pd.crosstab(df['Price range'], df['Has Table booking'], margins=True)

print("Cross-tabulation of price range vs. online delivery:")
print(cross_tab_online_delivery)

print("\nCross-tabulation of price range vs. table booking:")
print(cross_tab_table_booking)
```

```python
#We can observe the following relationships between price range and the availability of online delivery as well as table booking:

#Price Range vs. Online Delivery:

#Higher price range restaurants tend to have a higher proportion of online delivery available.
#As the price range increases, the number of restaurants offering online delivery decreases, except for price range 4 where the number is significantly lower compared to price range 3.
#The majority of restaurants in price range 1 and 2 offer online delivery, while fewer restaurants in price ranges 3 and 4 offer this service.

#Price Range vs. Table Booking:

#Price range 1 restaurants almost universally do not offer table booking, with only one exception.
#There's a clear trend where higher price range restaurants are more likely to offer table booking.
#Price ranges 3 and 4 have a similar proportion of restaurants offering table booking, but the total number of restaurants in price range 4 is significantly lower.

#These observations suggest that there is indeed a relationship between price range and the availability of online delivery as well as table booking. Higher-end restaurants (higher price range) tend to offer more amenities such as online delivery and table booking compared to lower-end restaurants.
```

```python
#Determine if higher-priced restaurants are more likely to offer these services.

# Calculating the proportion of restaurants offering online delivery within each price range
online_delivery_proportion = cross_tab_online_delivery.div(cross_tab_online_delivery['All'], axis=0)
print("Proportion of restaurants offering online delivery within each price range:")
print(online_delivery_proportion)

# Calculating the proportion of restaurants offering table booking within each price range
table_booking_proportion = cross_tab_table_booking.div(cross_tab_table_booking['All'], axis=0)
print("\nProportion of restaurants offering table booking within each price range:")
print(table_booking_proportion)
```

```python
#We can observe the following trends:

#Online Delivery:

#In price range 1, the majority of restaurants (84.23%) do not offer online delivery, while in higher price ranges, the proportion of restaurants offering online delivery decreases. Only around 9% of restaurants in price range 4 offer online delivery.
#There is a clear trend where higher-priced restaurants are less likely to offer online delivery.

#Table Booking:

#Price range 1 restaurants almost universally do not offer table booking, with only a negligible proportion offering this service.
#Higher-priced restaurants (price ranges 3 and 4) are more likely to offer table booking compared to lower-priced restaurants (price ranges 1 and 2). Over 45% of restaurants in price range 3 and 4 offer table booking.
#However, there's a slight decrease in the proportion of restaurants offering table booking from price range 2 to price range 4.

#These observations indicate that higher-priced restaurants are indeed more likely to offer table booking services, while they are less likely to offer online delivery services compared to lower-priced restaurants.
```
